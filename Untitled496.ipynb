```python
# ==========================================
# IMPORTS
# ==========================================
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# ==========================================
# PATH CONFIG
# ==========================================
BASE_PATH = r"C:\Users\NXTWAVE\Downloads\Justice Delay Analytics for Senior Citizen Cases (Karnataka HC)"
CSV_PATH = os.path.join(BASE_PATH, "casesdisposedofseniorcitizenHCKBengaluru_3.csv")

RESULT_CSV = os.path.join(BASE_PATH, "optimizer_comparison_results.csv")
BAR_GRAPH = os.path.join(BASE_PATH, "optimizer_comparison_bar.png")

# ==========================================
# LOAD & CLEAN DATA
# ==========================================
df = pd.read_csv(CSV_PATH)
df.columns = [c.strip().lower().replace(".", "").replace(" ", "_") for c in df.columns]

df = df[["month", "year", "count"]].dropna()
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["count"] = pd.to_numeric(df["count"], errors="coerce")
df = df.dropna()

month_map = {
    "January":1,"February":2,"March":3,"April":4,"May":5,"June":6,
    "July":7,"August":8,"September":9,"October":10,"November":11,"December":12
}
df["month"] = df["month"].astype(str).str.strip().str.title()
df["month_num"] = df["month"].map(month_map)

df.dropna(inplace=True)

X = df[["year", "month_num"]].values
y = df["count"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ==========================================
# FITNESS FUNCTION
# ==========================================
def evaluate_solution(neurons, lr):
    model = Sequential([
        Dense(neurons, activation="relu", input_shape=(2,)),
        Dense(neurons//2, activation="relu"),
        Dense(1)
    ])
    model.compile(optimizer=Adam(lr), loss="mse")
    model.fit(X_train, y_train, epochs=40, batch_size=8, verbose=0)
    preds = model.predict(X_test).flatten()
    return mean_squared_error(y_test, preds)

# ==========================================
# AIS
# ==========================================
def AIS():
    best = None
    best_score = float("inf")
    for _ in range(8):
        neurons = random.randint(8, 64)
        lr = random.uniform(0.001, 0.02)
        score = evaluate_solution(neurons, lr)
        if score < best_score:
            best_score = score
            best = (neurons, lr)
    return best_score

# ==========================================
# GA
# ==========================================
def GA():
    population = [(random.randint(8,64), random.uniform(0.001,0.02)) for _ in range(8)]
    for _ in range(4):
        population = sorted(population, key=lambda x: evaluate_solution(*x))
        parent1, parent2 = population[0], population[1]
        child = (
            int((parent1[0] + parent2[0]) / 2),
            (parent1[1] + parent2[1]) / 2
        )
        population[-1] = child
    return evaluate_solution(*population[0])

# ==========================================
# BA
# ==========================================
def BA():
    best_score = float("inf")
    for _ in range(8):
        neurons = random.randint(8,64)
        lr = random.uniform(0.001,0.02)
        score = evaluate_solution(neurons, lr)
        if score < best_score:
            best_score = score
    return best_score

# ==========================================
# ACO
# ==========================================
def ACO():
    best_score = float("inf")
    pheromone = 1.0
    for _ in range(8):
        neurons = random.randint(8,64)
        lr = random.uniform(0.001,0.02)
        score = evaluate_solution(neurons, lr)
        pheromone += 1/(score+1e-6)
        if score < best_score:
            best_score = score
    return best_score

# ==========================================
# RUN COMPARISON
# ==========================================
results = {
    "AIS": AIS(),
    "GA": GA(),
    "BA": BA(),
    "ACO": ACO()
}

# Convert to DataFrame
results_df = pd.DataFrame({
    "Optimizer": list(results.keys()),
    "MSE": list(results.values())
})

results_df["RMSE"] = np.sqrt(results_df["MSE"])

print("\nComparative Study Results:")
print(results_df)

results_df.to_csv(RESULT_CSV, index=False)

# ==========================================
# BAR GRAPH
# ==========================================
plt.figure(figsize=(8,5))
plt.bar(results_df["Optimizer"], results_df["RMSE"])
plt.title("Optimizer Comparison (RMSE)")
plt.ylabel("RMSE")
plt.tight_layout()
plt.show()
plt.savefig(BAR_GRAPH)
plt.close()

print("\nCOMPARATIVE STUDY COMPLETED ✅")
```
