```python
import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# ============================
# PATH CONFIG
# ============================
BASE_PATH = r"C:\Users\NXTWAVE\Downloads\Justice Delay Analytics for Senior Citizen Cases (Karnataka HC)"
CSV_PATH = os.path.join(BASE_PATH, "casesdisposedofseniorcitizenHCKBengaluru_3.csv")

RESULT_CSV = os.path.join(BASE_PATH, "aso_actual_vs_predicted.csv")
PREDICTION_CSV = os.path.join(BASE_PATH, "aso_future_predictions.csv")

LOSS_GRAPH = os.path.join(BASE_PATH, "aso_loss_curve.png")
HEATMAP_GRAPH = os.path.join(BASE_PATH, "aso_heatmap.png")
COMPARISON_GRAPH = os.path.join(BASE_PATH, "aso_actual_vs_predicted.png")
FORECAST_GRAPH = os.path.join(BASE_PATH, "aso_future_forecast.png")

# ============================
# LOAD & CLEAN DATA
# ============================
df = pd.read_csv(CSV_PATH)
df.columns = [c.strip().lower().replace(".", "").replace(" ", "_") for c in df.columns]

df = df[["month", "year", "count"]].dropna()
df["year"] = df["year"].astype(int)
df["count"] = df["count"].astype(float)

df["month"] = df["month"].astype(str).str.strip().str.title()

month_map = {
    "January":1,"February":2,"March":3,"April":4,"May":5,"June":6,
    "July":7,"August":8,"September":9,"October":10,"November":11,"December":12
}
df["month_num"] = df["month"].map(month_map)
df.dropna(inplace=True)
df["month_num"] = df["month_num"].astype(int)

print("\nDataset Preview:")
print(df.head())

# ============================
# HEATMAP (SHOW + SAVE)
# ============================
pivot = df.pivot_table(values="count", index="year", columns="month_num", aggfunc="sum")

plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap="YlOrRd", annot=True, fmt=".0f")
plt.title("ASO (AIS + ACO) Heatmap")
plt.show()
plt.savefig(HEATMAP_GRAPH)
plt.close()

# ============================
# FEATURE PREP
# ============================
X = df[["year", "month_num"]].values
y = df["count"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================
# AIS INITIAL POPULATION
# ============================
def ais_population(size=8):
    population = []
    for _ in range(size):
        population.append({
            "neurons": random.randint(8, 64),
            "lr": round(random.uniform(0.001, 0.02), 4)
        })
    return population

# ============================
# FITNESS FUNCTION
# ============================
def fitness(sol):
    model = Sequential([
        Dense(sol["neurons"], activation="relu", input_shape=(2,)),
        Dense(sol["neurons"] // 2, activation="relu"),
        Dense(1)
    ])
    model.compile(
        optimizer=Adam(learning_rate=sol["lr"]),
        loss="mse"
    )
    model.fit(X_train, y_train, epochs=50, batch_size=8, verbose=0)
    preds = model.predict(X_test).flatten()
    return mean_squared_error(y_test, preds)

# ============================
# ACO OPTIMIZATION
# ============================
def aco_optimize(population, iterations=6, alpha=1.0, rho=0.3):
    pheromone = [1.0 for _ in population]

    for _ in range(iterations):
        fitness_scores = [fitness(sol) for sol in population]

        # Update pheromone (inverse of error)
        for i in range(len(pheromone)):
            pheromone[i] = (1 - rho) * pheromone[i] + (1 / (fitness_scores[i] + 1e-6))

        # Probability distribution
        probs = np.array(pheromone) ** alpha
        probs = probs / probs.sum()

        # Ant-based sampling
        new_population = []
        for _ in population:
            idx = np.random.choice(len(population), p=probs)
            base = population[idx]

            new_solution = {
                "neurons": max(8, int(base["neurons"] + random.randint(-4, 4))),
                "lr": max(0.0005, round(base["lr"] + random.uniform(-0.002, 0.002), 4))
            }
            new_population.append(new_solution)

        population = new_population

    return min(population, key=fitness)

# ============================
# RUN AIS + ACO
# ============================
population = ais_population()
best_solution = aco_optimize(population)

print("\nBest ASO Solution:", best_solution)

# ============================
# FINAL MODEL
# ============================
model = Sequential([
    Dense(best_solution["neurons"], activation="relu", input_shape=(2,)),
    Dense(best_solution["neurons"] // 2, activation="relu"),
    Dense(1)
])

model.compile(
    optimizer=Adam(learning_rate=best_solution["lr"]),
    loss="mse"
)

early_stop = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=0
)

# ============================
# LOSS GRAPH
# ============================
plt.figure(figsize=(8,5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("ASO (AIS + ACO) Loss Curve")
plt.show()
plt.savefig(LOSS_GRAPH)
plt.close()

# ============================
# RESULTS
# ============================
y_pred = model.predict(X_test).flatten()

results = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})
results.to_csv(RESULT_CSV, index=False)

plt.figure(figsize=(8,5))
plt.plot(results["Actual"].values, label="Actual")
plt.plot(results["Predicted"].values, label="Predicted")
plt.legend()
plt.title("ASO Actual vs Predicted")
plt.show()
plt.savefig(COMPARISON_GRAPH)
plt.close()

# ============================
# FUTURE FORECAST
# ============================
last_year = df["year"].max()
future = pd.DataFrame({
    "year": [last_year + 1] * 12,
    "month_num": list(range(1, 13))
})

future_scaled = scaler.transform(future.values)
future_preds = model.predict(future_scaled).flatten()

future["predicted_count"] = future_preds
future.to_csv(PREDICTION_CSV, index=False)

plt.figure(figsize=(10,5))
plt.plot(future["month_num"], future["predicted_count"], marker="o")
plt.title("ASO (AIS + ACO) Future Forecast")
plt.xlabel("Month")
plt.ylabel("Predicted Cases")
plt.show()
plt.savefig(FORECAST_GRAPH)
plt.close()

print("\nASO (AIS + ACO) PIPELINE COMPLETED SUCCESSFULLY ")
```
